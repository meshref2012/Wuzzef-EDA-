# **Importing libaries**

```python
# import libaries 
import pandas as pd 
import numpy as np 
import seaborn as sns 
import matplotlib.pylab as plt
from wordcloud import WordCloud
```

# **loading data from kaggle**

```python
data = pd.read_csv('../input/wuzzuf-jobs/Wuzzuf_Jobs.csv')

data.head() #view the first 5 rows in data 
```

```python
# Check if data contains Missing Value Or NOT
data.isnull().sum()
```

```python
sns.heatmap(data.isna())   #showing null value in heatmap plot
plt.show()
```

```python
data.info()
```

```python
data.describe()
```

```python
# check for duplicate Value in Data 
x=data.duplicated().sum()
if(x==0):
    print("no duplicate in data")
else :
    print("There is",x, "Dublicate rows in data ")
```

```python
# Removing Dublicate rows in data 
data=data.drop_duplicates()
```

# **We Will Focous on Skills Column**

```python
data.columns
```

```python
print(data['Skills'].head())
```

```python
# splitting skills 
Skills_df = data.assign(Skills=data['Skills'].str.split(',')).explode('Skills')
Skills_df.head(10)
```

```python
# increase visualizations size 
plt.rcParams["figure.figsize"] = 9,6
# plotting  top 10 skills 
sns.countplot(data=Skills_df, y="Skills",order=Skills_df.Skills.value_counts().iloc[:10].index)
plt.title("Top 10 required skills ")
plt.show()
```

the previous plot show that English speaking is highly required skill in Wazuf , Second sales skills , third the computer knowledge. We can note in general that the most frequency required skills related with Tecknology skills , sales , marketing , customers support and finance.

```python
# Plotting the TOP 20 Jops location 
sns.countplot(data = data, y = "Location",order=data.Location.value_counts().iloc[:20].index );
plt.title("Top 20 gobs Locations")
```

We notes That Most Countries Require people For Working is Cairo The Capital of egypt .

```python
# Top required experiences levels
sns.countplot( data = data , x = "Level" , order = data.Level.value_counts().iloc[:].index) #plot all level 
plt.title("Top required experiences levels")
plt.xticks(rotation= 70)
plt.show()
```

We note from the previous plot
Companies prefer to hire people with a lot of experience
, and the opportunity for students to be employed is very few.

# **Word Cloud of Skills Required**

```python
job_title=data['Skills']
wordCloud = WordCloud(width=450,height= 300).generate(' '.join(job_title))
plt.figure(figsize=(19,9))
plt.axis('off')
plt.title('Most common skills required',fontsize=20)
plt.imshow(wordCloud)
plt.show()
```
